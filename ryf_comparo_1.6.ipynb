```python
import xarray as xr
```

```python
from matplotlib import pyplot as plt
```

```python
from dask.distributed import Client
```

```python
from numpy.testing import assert_allclose
```

```python
client = Client(threads_per_worker = 1, n_workers=12)
client
```

```python
for iVar in [
        "rsds",
        "rlds",
        "prra",
        "prsn",
        "psl",
        "friver",
        "tas",
        "huss",
        "uas",
        "vas",
        "licalvf",
    ] :
    print(iVar)
    new_ds = xr.open_dataset(f'/g/data/tm70/as2285/om3-scripts/make_ryf/RYF.{iVar}.1990_1991.nc', chunks='auto')
    old_ds = xr.open_dataset(f'/g/data/vk83/experiments/inputs/JRA-55/RYF/v1-4/data/RYF.{iVar}.1990_1991.nc', chunks='auto')
    print(new_ds.time)
    new_ds[iVar].sum(['lat','lon']).plot()
    old_ds[iVar].sum(['lat','lon']).plot()
    plt.show()
    plt.pcolor(new_ds[iVar].isel(time=0).values-old_ds[iVar].isel(time=0).values)
    plt.colorbar()
    plt.show()
    assert_allclose(new_ds[iVar].values, old_ds[iVar].values)
```
