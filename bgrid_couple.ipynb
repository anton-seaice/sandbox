Pick a var on the A grid:

```python
ds.SSH.rename({'xh':'lon','yh':'lat'})
```

Pick a the var of interest on the C grid:

```python
ds.SSU.rename({'xq':'lon','yh':'lat'})
```

```python
regridder = xesmf.Regridder(
    ds.SSU.rename({'xq':'lon','yh':'lat'}), ds.SSH.rename({'xh':'lon','yh':'lat'}), 'bilinear'
)
```

```python
plt.pcolor(
    (regridder(ds.SSU).values-cpl_1_ds['ocnImp_So_u'].values)[0,0:240,:],
    vmin=-0.03, vmax = 0.03
)
plt.colorbar()
```

```python
plt.hist(
    (regridder(ds.SSU).values-cpl_1_ds['ocnImp_So_u'].values)[0,0:240,:]
)
```
